```python


submission_list = [w2v7894, w2v_7932, w2v_7913, w2v_7883]

sum_pred = np.zeros(2482)
for sub in submission_list:
    
    sum_pred += sub.iloc[:,1] ** 3.5
    
pred = (sum_pred / len(submission_list)) ** (1 / 3.5)


fname = 'w2v7894, w2v_7932, w2v_7913, w2v_7883.csv'
IDtest = range(3500, 5982)
submissions = pd.concat([pd.Series(IDtest, name="cust_id"), pd.Series(pred, name="gender")] ,axis=1)
submissions.to_csv(fname, index=False)
print("'{}' is ready to submit." .format(fname))

pd.read_csv(fname, encoding='cp949')
```
